```python
from moviepy.editor import VideoFileClip
import os
from openai import OpenAI

# Step 1: Video to Audio Conversion
def convert_video_to_audio(video_path, audio_path):
    try:
        # Load video clip
        video_clip = VideoFileClip(video_path)

        # Extract audio
        audio_clip = video_clip.audio

        # Write audio file
        audio_clip.write_audiofile(audio_path)

        # Close video clip
        video_clip.close()

    except Exception as e:
        print(f"Error during video to audio conversion: {e}")
        

# Step 2: Transcription with LLM

# Modal OpenAI GPT-3
def transcribe_audio_with_llm_gpt(audio_path):
    try:
        # Set up OpenAI API key
        openai.api_key = "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"

        # Use OpenAI GPT-3 for transcription
        response = openai.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": f"Transcribe the following audio:\n{audio_path}"}
            ]
        )

        transcribed_text = response.choices[0].message.content
        return transcribed_text

    except Exception as e:
        print(f"Error during transcription: {e}")
        return None



# Modal OpenAI Whisper - 1

def transcribe_audio_with_llm_whisper(audio_path):
   
    openai.api_key = "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"
    audio_file = open(audio_path, "rb")
    transcript = openai.audio.transcriptions.create(
        model="whisper-1", 
        file=audio_file, 
        response_format="text"
    )
    print("whisper text", transcript)
    return transcript



# Step 3: Save Transcribed Text
def save_transcribed_text_gpt(transcribed_text, output_path):
    try:
        with open(output_path, 'w') as file:
            file.write(transcribed_text)

    except Exception as e:
        print(f"Error during saving transcribed text: {e}")



def save_transcribed_text_whisper(transcribed_text, output_path):
    try:
        with open(output_path, 'w') as file:
            file.write(transcribed_text)

    except Exception as e:
        print(f"Error during saving transcribed text: {e}")



# Example Usage
video_path = '/Users/pujakumari/Downloads/cosmos_original.mp4'
audio_path = '/Users/pujakumari/Downloads/audio.wav'
output_text_path_gpt = '/Users/pujakumari/Downloads/outputVideo_gpt.txt'
output_text_path_whisper = '/Users/pujakumari/Downloads/outputVideo_whisper.txt'

# Step 1: Convert video to audio
convert_video_to_audio(video_path, audio_path)

# Step 2: Transcribe audio with LLM
transcribed_text_whisper = transcribe_audio_with_llm_whisper(audio_path)
transcribed_text_gpt = transcribe_audio_with_llm_gpt(audio_path)


# Step 3: Save transcribed text
if transcribed_text_gpt is not None:
    save_transcribed_text_gpt(transcribed_text_gpt, output_text_path_gpt)

if transcribed_text_whisper is not None:
    save_transcribed_text_whisper(transcribed_text_whisper, output_text_path_whisper)
```
